#### Loading example and input file

```python
with open('example1.txt') as input_file:
    lines_ex1 = input_file.readlines()

# with open('example2.txt') as input_file:
#     lines_ex2 = input_file.read()

with open('input.txt') as input_file:
    lines = input_file.readlines()
```

#### Common functions

```python
def parse(lines):
	ranges = []
	ids = []
	
	i = 0
	while lines[i] != '\n':
		range = lines[i].strip().split('-')
		ranges.append((int(range[0]), int(range[1])))
		i += 1
	i += 1
	while i < len(lines):
		ids.append(int(lines[i].strip()))
		i += 1

	return ranges, ids
```

#### Part One

```python
def part_one(input_lines):
	ranges, ids = parse(input_lines)
	fresh = set()

	for id in ids:
		for range in ranges:
			if range[0] <= id <= range[1]:
				fresh.add(id)
				break
	
	return len(fresh)

print("Example input: " + str(part_one(lines_ex1)))
print("Real input: " + str(part_one(lines)))
```

#### Part Two

```python
def part_two(input_lines):
	id_ranges, _ = parse(input_lines)
	fresh = 0
	id_ranges.sort()

	i = 0
	while i < len(id_ranges):
		j = i + 1
		while j < len(id_ranges):
			if id_ranges[i][1] >= id_ranges[j][0]:
				new_range = (id_ranges[i][0], max(id_ranges[i][1], id_ranges[j][1]))
				id_ranges[i] = new_range
				id_ranges.pop(j)
			else:
				j += 1
		i += 1
	
	for r in id_ranges:
		fresh += r[1] - r[0] + 1

	return fresh

print("Example input: " + str(part_two(lines_ex1)))
print("Real input: " + str(part_two(lines)))
```
