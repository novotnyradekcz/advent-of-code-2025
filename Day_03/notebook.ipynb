#### Loading example and input file

```python
with open('example1.txt') as inputFile:
    linesEx1 = inputFile.readlines()

# with open('example2.txt') as inputFile:
#     linesEx2 = inputFile.read()

with open('input.txt') as inputFile:
    lines = inputFile.readlines()
```

#### Common functions

```python
def parse(input):
	banks = []
	for line in input:
		bank = []
		for char in line.strip():
			bank.append(int(char))
		banks.append(bank)
	return banks
```

#### Part One

```python
def part_one(inputLines):
	result = 0
	banks = parse(inputLines)

	for bank in banks:
		first = max(bank[:-1])
		second = max(bank[bank.index(first) + 1:])
		result += int(str(first) + str(second))

	return result

print("Example input: " + str(part_one(linesEx1)))
print("Real input: " + str(part_one(lines)))
```

#### Part Two

```python
def part_two(inputLines):
	result = 0
	banks = parse(inputLines)

	for bank in banks:
		digits = []

		digits.append(max(bank[:-11]))
		for i in range(10):
			bank = bank[bank.index(digits[-1]) + 1:]
			digit = max(bank[:-10 + i])
			digits.append(digit)
		digits.append(max(bank[bank.index(digits[-1]) + 1:]))

		joltage = ""
		for digit in digits:
			joltage += str(digit)

		result += int(joltage)

	return result

print("Example input: " + str(part_two(linesEx1)))
print("Real input: " + str(part_two(lines)))
```
