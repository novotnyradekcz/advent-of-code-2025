#### Common functions

#### Part One

```python
def parse_one(lines):
	problems = []

	for i in range(len(lines) - 1):
		problem = lines[i].strip().split()
		for j in range(len(problem)):
			problem[j] = int(problem[j])
		problems.append(problem)
	problems.append(lines[-1].strip().split())

	return problems

def part_one(input_lines):
	problems = parse_one(input_lines)
	total = 0

	for i in range(len(problems[0])):
		if problems[-1][i] == '+':
			result = problems[0][i]
			for j in range(1, len(problems) - 1):
				result += problems[j][i]
		elif problems[-1][i] == '*':
			result = problems[0][i]
			for j in range(1, len(problems) - 1):
				result *= problems[j][i]
		total += result

	return total

print("Example input: " + str(part_one(lines_ex1)))
print("Real input: " + str(part_one(lines)))
```

#### Part Two

```python
def parse_two(lines):
	problems = []

	for i in range(len(lines[0])):
		if lines[-1][i] == '+' or lines[-1][i] == '*':
			problems.append(lines[-1][i])
		number = ""
		for j in range(len(lines) - 1):
			number += lines[j][i]
		if number.strip() != "":
			problems.append(int(number))

	return problems

def part_two(input_lines):
	problems = parse_two(input_lines)
	total = 0
	i = 0
	
	while i < len(problems):
		if problems[i] == '+':
			i += 1
			result = problems[i]
			i += 1
			while i < len(problems) and problems[i] != '+' and problems[i] != '*':
				result += problems[i]
				i += 1
			total += result
		elif problems[i] == '*':
			i += 1
			result = problems[i]
			i += 1
			while i < len(problems) and problems[i] != '+' and problems[i] != '*':
				result *= problems[i]
				i += 1
			total += result
			
	return total
	

print("Example input: " + str(part_two(lines_ex1)))
print("Real input: " + str(part_two(lines)))
```
