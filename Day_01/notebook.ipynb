#### Loading example and input file

```python
with open('example1.txt') as inputFile:
    linesEx1 = inputFile.readlines()

# with open('example2.txt') as inputFile:
#     linesEx2 = inputFile.readlines()

with open('input.txt') as inputFile:
    lines = inputFile.readlines()
```

#### Common functions

```python
def parse(input):
	rotations = []
	
	for line in input:
		if line[0] == 'L':
			direction = -1
		else:
			direction = 1
		rotations.append((direction * int(line[1:])))

	return rotations
```

#### Part One

```python
def part_one(inputLines):
	rotations = parse(inputLines)
	zeroes = 0
	current = 50

	for rotation in rotations:
		current = (current + rotation) % 100
		if current == 0:
			zeroes += 1
	
	return zeroes


print("Example input: " + str(part_one(linesEx1)))
print("Real input: " + str(part_one(lines)))
```

#### Part Two

```python
def part_two(inputLines):
	rotations = parse(inputLines)
	zeroes = 0
	current = 50

	for rotation in rotations:
		for step in range(abs(rotation)):
			if rotation > 0:
				current += 1
			else:
				current -= 1
			current = current % 100
			if current == 0:
				zeroes += 1
	
	return zeroes

print("Example input: " + str(part_two(linesEx1)))
print("Real input: " + str(part_two(lines)))
```
