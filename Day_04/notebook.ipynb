#### Loading example and input file

```python
with open('example1.txt') as inputFile:
    linesEx1 = inputFile.readlines()

# with open('example2.txt') as inputFile:
#     linesEx2 = inputFile.read()

with open('input.txt') as inputFile:
    lines = inputFile.readlines()
```

#### Common functions

```python
def parse(input):
	coils = []
	for row in input:
		coilRow = []
		for coil in row.strip():
			coilRow.append(coil)
		coils.append(coilRow)
	return coils
```

#### Part One

```python
def part_one(inputLines):
	coils = parse(inputLines)
	
	accessible = 0
	
	for i in range(len(coils)):
		for j in range(len(coils[i])):
			if coils[i][j] == '@':
				neighbours = 0
				if i > 0 and coils[i - 1][j] == '@':
					neighbours += 1
				if i > 0 and j > 0 and coils[i - 1][j - 1] == '@':
					neighbours += 1
				if i > 0 and j < len(coils[i]) - 1 and coils[i - 1][j + 1] == '@':
					neighbours += 1
				if i < len(coils) - 1 and coils[i + 1][j] == '@':
					neighbours += 1
				if i < len(coils) - 1 and j > 0 and coils[i + 1][j - 1] == '@':
					neighbours += 1
				if i < len(coils) - 1 and j < len(coils[i]) - 1 and coils[i + 1][j + 1] == '@':
					neighbours += 1
				if j > 0 and coils[i][j - 1] == '@':
					neighbours += 1
				if j < len(coils[i]) - 1 and coils[i][j + 1] == '@':
					neighbours += 1
				if neighbours < 4:
					accessible += 1

	return accessible
				
	

print("Example input: " + str(part_one(linesEx1)))
print("Real input: " + str(part_one(lines)))
```

#### Part Two

```python
def part_two(inputLines):
	coils = parse(inputLines)
	removed = 0
	wasRemoved = True

	while wasRemoved:
		wasRemoved = False
		for i in range(len(coils)):
			for j in range(len(coils[i])):
				if coils[i][j] == '@':
					neighbours = 0
					if i > 0 and coils[i - 1][j] == '@':
						neighbours += 1
					if i > 0 and j > 0 and coils[i - 1][j - 1] == '@':
						neighbours += 1
					if i > 0 and j < len(coils[i]) - 1 and coils[i - 1][j + 1] == '@':
						neighbours += 1
					if i < len(coils) - 1 and coils[i + 1][j] == '@':
						neighbours += 1
					if i < len(coils) - 1 and j > 0 and coils[i + 1][j - 1] == '@':
						neighbours += 1
					if i < len(coils) - 1 and j < len(coils[i]) - 1 and coils[i + 1][j + 1] == '@':
						neighbours += 1
					if j > 0 and coils[i][j - 1] == '@':
						neighbours += 1
					if j < len(coils[i]) - 1 and coils[i][j + 1] == '@':
						neighbours += 1
					if neighbours < 4:
						coils[i][j] = '.'
						removed += 1
						wasRemoved = True

	return removed

print("Example input: " + str(part_two(linesEx1)))
print("Real input: " + str(part_two(lines)))
```
