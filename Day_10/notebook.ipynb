#### Part Two

```python
def parse_two(lines):
	buttons = []
	joltages = []
	
	for line in lines:
		parts = line.strip().split(' ')

		joltage_string = parts[-1].strip('{}').split(',')
		joltage = [int(x) for x in joltage_string]
		joltages.append(np.array(joltage))

		button_list = []
		for i in range(1, len(parts) - 1):
			button_string = parts[i].strip('()').split(',')
			button = tuple([int(x) for x in button_string])
			button_array = np.zeros((len(joltage),), dtype=int)
			for i in range(len(joltage)):
				if i in button:
					button_array[i] = 1
			button_list.append(button_array)
		buttons.append(np.array(button_list))

	return buttons, joltages

def part_two(input_lines):
	buttons, joltages = parse_two(input_lines)

	button_presses = 0
	for i in range(len(buttons)):
		basis = buttons[i].T
		target = joltages[i].T
		n_features, n_basis = basis.shape
		coefficients = [pulp.LpVariable(f"c{i}", lowBound=0, cat="Integer") for i in range(n_basis)]

		problem = pulp.LpProblem("Minimize_Coefficient_Sum", pulp.LpMinimize)
		problem += pulp.lpSum(coefficients)
		for row in range(n_features):
			problem += pulp.lpSum(basis[row, col] * coefficients[col] for col in range(n_basis)) == target[row]
		problem.solve(pulp.PULP_CBC_CMD(msg=False))

		button_presses += sum([int(v.value()) for v in coefficients])
		
	return button_presses

print("Example input: " + str(part_two(lines_ex1)))
print("Real input: " + str(part_two(lines)))
```
