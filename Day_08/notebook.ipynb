#### Common functions

```python
def parse(lines):
	boxes = []
	for line in lines:
		box = line.strip().split(',')
		boxes.append(tuple(int(x) for x in box))
	return boxes

def distance(a, b):
	return abs(a[0] - b[0])**2 + abs(a[1] - b[1])**2 + abs(a[2] - b[2])**2
```

#### Part One

```python
def part_one(input_lines, n_connections=1000):
	boxes = parse(input_lines)

	distances = {}
	for i in range(len(boxes)):
		for j in range(i + 1, len(boxes)):
			dist = distance(boxes[i], boxes[j])
			distances[(i, j)] = dist
	distances_sorted = sorted(distances.items(), key=lambda x: x[1])

	circuits = []
	counter = 0
	for (a, b), dist in distances_sorted:
		counter += 1
		if counter > n_connections:
			break

		added = False
		for i in range(len(circuits)):
			if a in circuits[i] or b in circuits[i]:
				circuits[i].add(a)
				circuits[i].add(b)
				added = True

		if not added:
			circuits.append(set([a, b]))
	
		for i in range(len(circuits)):
			for j in range(i + 1, len(circuits)):
				if len(circuits[i].intersection(circuits[j])) > 0:
					circuits[i] = circuits[i].union(circuits[j])
					circuits[j] = set()

	sorted_circuits = sorted(circuits, key=lambda x: len(x), reverse=True)
	return len(sorted_circuits[0]) * len(sorted_circuits[1]) * len(sorted_circuits[2])

print("Example input: " + str(part_one(lines_ex1, 10)))
print("Real input: " + str(part_one(lines)))
```

#### Part Two

```python
def part_two(input_lines):
	boxes = parse(input_lines)

	distances = {}
	for i in range(len(boxes)):
		for j in range(i + 1, len(boxes)):
			dist = distance(boxes[i], boxes[j])
			distances[(i, j)] = dist
	distances_sorted = sorted(distances.items(), key=lambda x: x[1])

	circuits = []
	for (a, b), dist in distances_sorted:
		circuits.append(set([a, b]))
	
		for i in range(len(circuits)):
			for j in range(i + 1, len(circuits)):
				if len(circuits[i].intersection(circuits[j])) > 0:
					circuits[i] = circuits[i].union(circuits[j])
					circuits[j] = set()
		
		circuits = list(filter(lambda x: len(x) > 0, circuits))
		if len(circuits) == 1 and len(circuits[0]) == len(boxes):
			break
		
	return boxes[a][0] * boxes[b][0]

print("Example input: " + str(part_two(lines_ex1)))
print("Real input: " + str(part_two(lines)))
```
