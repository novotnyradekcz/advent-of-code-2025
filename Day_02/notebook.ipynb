#### Loading example and input file

```python
with open('example1.txt') as input_file:
    lines_ex1 = input_file.read()

# with open('example2.txt') as input_file:
#     lines_ex2 = input_file.read()

with open('input.txt') as input_file:
    lines = input_file.read()
```

#### Common functions

```python
def parse(input):
	ranges = []
	
	range_strings = input.split(',')
	for range_string in range_strings:
		bounds = range_string.split('-')
		ranges.append((int(bounds[0]), int(bounds[1])))
	
	return ranges
```

#### Part One

```python
def part_one(input_lines):
	ranges = parse(input_lines)
	invalid = []
	
	for interval in ranges:
		for number in range(interval[0], interval[1] + 1):
			str_number = str(number)
			if len(str_number) % 2 == 1:
				continue
			if str_number[:len(str_number)//2] == str_number[len(str_number)//2:]:
				invalid.append(number)

	return sum(invalid)

print("Example input: " + str(part_one(lines_ex1)))
print("Real input: " + str(part_one(lines)))
```

#### Part Two

```python
def part_two(input_lines):
	ranges = parse(input_lines)
	invalid = set()
	
	for interval in ranges:
		for number in range(interval[0], interval[1] + 1):
			str_number = str(number)
			
			for i in range(2, len(str_number) + 1):
				if len(str_number) % i == 0:
					part = len(str_number) // i;
					is_invalid = True
					for j in range(1, i):
						if str_number[part * (j - 1): part * j] != str_number[part * j: part * (j + 1)]:
							is_invalid = False
							break
					if is_invalid:
						invalid.add(number)

	return sum(invalid)

print("Example input: " + str(part_two(lines_ex1)))
print("Real input: " + str(part_two(lines)))
```
