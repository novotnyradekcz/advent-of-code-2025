#### Loading example and input file

```python
with open('example1.txt') as inputFile:
    linesEx1 = inputFile.read()

# with open('example2.txt') as inputFile:
#     linesEx2 = inputFile.readlines()

with open('input.txt') as inputFile:
    lines = inputFile.read()
```

#### Common functions

```python
def parse(input):
	ranges = []
	
	rangeStrings = input.split(',')
	for rangeString in rangeStrings:
		bounds = rangeString.split('-')
		ranges.append((int(bounds[0]), int(bounds[1])))
	
	return ranges
```

#### Part One

```python
def part_one(inputLines):
	ranges = parse(inputLines)
	invalid = []
	
	for interval in ranges:
		for number in range(interval[0], interval[1] + 1):
			strNumber = str(number)
			if len(strNumber) % 2 == 1:
				continue
			if strNumber[:len(strNumber)//2] == strNumber[len(strNumber)//2:]:
				invalid.append(number)

	return sum(invalid)

print("Example input: " + str(part_one(linesEx1)))
print("Real input: " + str(part_one(lines)))
```

#### Part Two

```python
def part_two(inputLines):
	ranges = parse(inputLines)
	invalid = set()
	
	for interval in ranges:
		for number in range(interval[0], interval[1] + 1):
			strNumber = str(number)
			
			for i in range(2, len(strNumber) + 1):
				if len(strNumber) % i == 0:
					part = len(strNumber) // i;
					isInvalid = True
					for j in range(1, i):
						if strNumber[part * (j - 1): part * j] != strNumber[part * j: part * (j + 1)]:
							isInvalid = False
							break
					if isInvalid:
						invalid.add(number)

	return sum(invalid)

print("Example input: " + str(part_two(linesEx1)))
print("Real input: " + str(part_two(lines)))
```
