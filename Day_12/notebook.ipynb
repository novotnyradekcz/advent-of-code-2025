#### Common functions

```python
def parse(input_lines):
	i = 0
	shapes = {}
	region_sizes = {}
	region_shapes = {}

	while "x" not in input_lines[i]:
		num_index = int(input_lines[i].strip(':\n'))
		i += 1
		tiles = set()
		x = 0
		while input_lines[i].strip() != "":
			row = input_lines[i].strip()
			for y in range(len(row)):
				if row[y] == "#":
					tiles.add((x, y))
			x += 1
			i += 1
		shapes[num_index] = tiles
		i += 1
	
	j = 0
	while i < len(input_lines):
		parts = input_lines[i].strip().split(':')
		region_size = parts[0].split('x')
		region_sizes[j] = (int(region_size[0]), int(region_size[1]))
		shape_counts = parts[1].strip().split()
		region_shapes[j] = [int(count) for count in shape_counts]
		i += 1
		j += 1

	return shapes, region_sizes, region_shapes
```

#### Part One

```python
def part_one(input_lines):
	shapes, region_sizes, region_shapes = parse(input_lines)

	larger_than_upper = 0
	smaller_than_lower = 0
	inbetween = 0
	for i, shape_counts in region_shapes.items():
		upper_bound = 0
		lower_bound = 0
		for j, shape_count in enumerate(shape_counts):
			upper_bound += (max(shapes[j], key=lambda x: x[0])[0] + 1) * (max(shapes[j], key=lambda x: x[1])[1] + 1) * shape_count
			lower_bound += len(shapes[j]) * shape_count
		if lower_bound < region_sizes[i][0] * region_sizes[i][1] < upper_bound:
			inbetween += 1
		# the following comparison is actually incorrect and shouldn;t work
		# e. g. if we want to fit four 3x3 shapes into a 4x9 region, the area comparison would say it's possible
		# but in reality it's not possible due to shape arrangement
		# however, I seem to have gotten lucky with my input
		elif region_sizes[i][0] * region_sizes[i][1] >= upper_bound:
			larger_than_upper += 1
		else:
			smaller_than_lower += 1
		
	# what the results mean:
	# number of presents that definitely won't fit, number of presents that may or my not fit depending on shapes, number of presents that definitely will fit
	# (even though this last number is actually wrong as explained above)
	# if the second number is zero, the third is the result (except it should not be, as explained above, I just got lucky)
	# if the second is non-zero, those lines need to be checked manually (good luck)
	return smaller_than_lower, inbetween, larger_than_upper

print("Example1 input: " + str(part_one(lines_ex1)))
print("Real input: " + str(part_one(lines)))
```
