### Day 9: Movie Theater

#### Importing libraries

```python
import numpy as np
from shapely.geometry import Point, Polygon, box
```

#### Loading example and input file

```python
with open('example1.txt') as input_file:
    lines_ex1 = input_file.readlines()

# with open('example2.txt') as inputFile:
#     linesEx2 = inputFile.read()

with open('input.txt') as input_file:
    lines = input_file.readlines()
```

#### Common functions

```python
def parse(lines):
	red_tiles = []
	for line in lines:
		tile = line.strip().split(',')
		red_tiles.append((int(tile[0]), int(tile[1])))
	
	return red_tiles
```

#### Part One

```python
def part_one(input_lines):
	red_tiles = parse(input_lines)
	largest = 0
	for i in range(len(red_tiles)):
		for j in range(i + 1, len(red_tiles)):
			area = (abs(red_tiles[i][0] - red_tiles[j][0]) + 1) * (abs(red_tiles[i][1] - red_tiles[j][1]) + 1)
			if area > largest:
				largest = area
		
	return largest

print("Example input: " + str(part_one(lines_ex1)))
print("Real input: " + str(part_one(lines)))
```

#### Part Two

```python
def part_two(input_lines):
	red_tiles = parse(input_lines)

	polygon = Polygon(red_tiles)
	largest_area = 0
	for i in range(len(red_tiles) - 1):
		for j in range(i + 1, len(red_tiles)):
			rectangle = box(min(red_tiles[i][0], red_tiles[j][0]), min(red_tiles[i][1], red_tiles[j][1]), max(red_tiles[i][0], red_tiles[j][0]), max(red_tiles[i][1], red_tiles[j][1]))
			if polygon.contains(rectangle):
				area = (abs(red_tiles[i][0] - red_tiles[j][0]) + 1) * (abs(red_tiles[i][1] - red_tiles[j][1]) + 1)
				if area > largest_area:
					largest_area = area

	return largest_area

print("Example input: " + str(part_two(lines_ex1)))
print("Real input: " + str(part_two(lines)))
```
